```python
df_train_labels = df_train['SalePrice']
```

```python
df_train.drop(['SalePrice'], axis=1, inplace=True)
```

```python
df_train = df_train.reindex(sorted(df_train.columns), axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

df_train_cols = df_train.columns
scaler = StandardScaler()
scaler.fit(df_train)
df_train = scaler.transform(df_train)
```

```python
df_train = pd.DataFrame(df_train, columns=df_train_cols)
df_train
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(df_train, df_train_labels)
```

```python
print("Predictions:", lin_reg.predict(df_train))
print("Labels:", df_train_labels.values)
```

```python
from sklearn.metrics import mean_squared_error
predictions = lin_reg.predict(df_train)
lin_mse = mean_squared_error(np.log(df_train_labels), np.log(predictions))
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
from sklearn.model_selection import cross_val_score
lin_scores = cross_val_score(lin_reg, df_train, df_train_labels,
                         scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-lin_scores)
display_scores(lin_rmse_scores)
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor()
tree_reg.fit(df_train, df_train_labels)
```

```python
print("Predictions:", tree_reg.predict(df_train))
print("Labels:", df_train_labels.values)
```

```python
predictions = tree_reg.predict(df_train)
tree_mse = mean_squared_error(np.log(df_train_labels), np.log(predictions))
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```

```python
tree_scores = cross_val_score(tree_reg, df_train, df_train_labels,
                         scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-tree_scores)
display_scores(tree_rmse_scores)
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor()
forest_reg.fit(df_train, df_train_labels)
```

```python
print("Predictions:", forest_reg.predict(df_train))
print("Labels:", df_train_labels.values)
```

```python
predictions = forest_reg.predict(df_train)
forest_mse = mean_squared_error(np.log(df_train_labels), np.log(predictions))
forest_rmse = np.sqrt(forest_mse)
forest_rmse
```

```python
forest_scores = cross_val_score(forest_reg, df_train, df_train_labels,
                         scoring="neg_mean_squared_error", cv=10)
forest_rmse_scores = np.sqrt(-forest_scores)
display_scores(forest_rmse_scores)
```
